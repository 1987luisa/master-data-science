```python
from sklearn.datasets import load_digits
data = load_digits()
```

```python
X=data.data
y=data.target
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=45)
```

```python
X_enc=tsne.fit_transform(X)
```
