```python
#Creamos DataFrames desde Diccionarios con funciones de distribución normal aleatoria:

const_ruido=pd.DataFrame.from_dict({'modul_precio':np.random.normal(750,220,1000), 
                        "retrasos_sems":np.random.normal(36,20,1000),
                        "duracion_sems":np.random.normal(65,31,1000),
                        "superf_const": np.random.normal(5800,3300,1000)})


const_ruido2=pd.DataFrame.from_dict({'modul_precio':np.random.normal(810,275,500), 
                        "retrasos_sems":np.random.normal(19,11,500),
                        "duracion_sems":np.random.normal(65,31,500),
                        "superf_const": np.random.normal(3800,2029,500)})


const_ruido3=pd.DataFrame.from_dict({'modul_precio':np.random.normal(750,220,200), 
                        "retrasos_sems":np.random.normal(36,20,200),
                        "duracion_sems":np.random.normal(63,20,200),
                        "superf_const": np.random.normal(8100,2200,200)})

const_ruido4=pd.DataFrame.from_dict({'modul_precio':np.random.normal(520,193,1200), 
                        "retrasos_sems":np.random.normal(6,4,1200),
                        "duracion_sems":np.random.normal(62,20,1200),
                        "superf_const": np.random.normal(8300,2281,1200)})


const_tot = pd.concat([const_ruido2 , const_ruido,const_ruido3,const_ruido4])

#Quitamos los valores menores de cero y otros absurdos.
const_tot = const_tot[(const_tot["duracion_sems"]>0 )& (const_tot['superf_const']>50)                       
                      & (const_tot['modul_precio']>181) & (const_tot['retrasos_sems']>=0)]

print(const_tot.describe())

const_tot.plot.scatter(x='duracion_sems', y='retrasos_sems', title="restrasos VS mod.Precio", alpha=0.1)
const_tot.plot.scatter(x='superf_const', y='retrasos_sems', title="restrasos VS mod.Precio", alpha=0.1)
const_tot.plot.scatter(x='modul_precio', y='retrasos_sems', title="restrasos VS mod.Precio", alpha=0.1)
const_tot.plot.scatter(x='duracion_sems', y='modul_precio', title="restrasos VS mod.Precio", alpha=0.1)
```
