```python
import pandas as pd
import random as rnd
import numpy as np

%matplotlib inline
```

# Creación de DB de la construcción:

```python
#Creamos DataFrames desde Diccionarios con funciones de distribución normal aleatoria:

cons0_aisl=pd.DataFrame.from_dict({'modul_precio':np.random.normal(910,375,600), 
                        "retrasos_sems":np.random.normal(7,22,600),
                        "duracion_sems":np.random.normal(30,20,600),
                        "superf_const": np.random.normal(125,350,600)})

cons1_med=pd.DataFrame.from_dict({'modul_precio':np.random.normal(750,220,1000), 
                        "retrasos_sems":np.random.normal(36,20,1000),
                        "duracion_sems":np.random.normal(65,31,1000),
                        "superf_const": np.random.normal(5800,3300,1000)})


cons2_peq=pd.DataFrame.from_dict({'modul_precio':np.random.normal(810,275,500), 
                        "retrasos_sems":np.random.normal(19,11,500),
                        "duracion_sems":np.random.normal(65,31,500),
                        "superf_const": np.random.normal(3800,2029,500)})


cons3_gran=pd.DataFrame.from_dict({'modul_precio':np.random.normal(600,220,200), 
                        "retrasos_sems":np.random.normal(10,15,200),
                        "duracion_sems":np.random.normal(63,20,200),
                        "superf_const": np.random.normal(8100,2200,200)})

cons4_gran=pd.DataFrame.from_dict({'modul_precio':np.random.normal(720,193,1200), 
                        "retrasos_sems":np.random.normal(23,10,1200),
                        "duracion_sems":np.random.normal(62,20,1200),
                        "superf_const": np.random.normal(8300,2281,1200)})


const_tot = pd.concat([cons0_aisl,cons1_med, cons2_peq,cons3_gran,cons4_gran])

#Quitamos los valores menores de cero y otros absurdos.
const_tot = const_tot[(const_tot["duracion_sems"]>0 )& (const_tot['superf_const']>50)                       
                      & (const_tot['modul_precio']>181) & (const_tot['retrasos_sems']>=0)]

const_tot=const_tot.reset_index().drop("index",axis=1)
                                  
const_tot.head(1000)
```

```python
list(np.unique(const_tot["tipolog"]))
```

```python
viv_aisl=[1 if n =="VIV_AISL" else 0 for n in const_tot["tipolog"]]
viv_col=[1 if n =="VIV_COL" else 0 for n in const_tot["tipolog"]]
otros=[1 if n =="OTROS" else 0 for n in const_tot["tipolog"]]

const_tot["VIV_AISL"]  =pd.Series(viv_aisl)
const_tot["VIV_COL"] =pd.Series(viv_col)
const_tot["OTROS"]=pd.Series(otros)

totaltipos= pd.DataFrame(pd.Series(data=(const_tot["VIV_AISL"].sum(),
                                   const_tot["VIV_COL"].sum(),
                                   const_tot["OTROS"].sum())))


totaltipos
```

```python
#Añadimos tipología de edificación como columna:

## Supervised randomizer:

ran90=[True,True,True,True,True,True,True,True,True,False]
ran80=[True,True,True,True,True,True,True,True,False,False]
ran60=[True,True,True,True,True,True,False,False,False,False]

edif=["VIV_AISL" if (n <250) and (True in rnd.sample(ran90,1)) else      
      "VIV_AISL" if (250<n<450) and (True in rnd.sample(ran80,1)) else
      "VIV_COL" if (450<n<3200) and (True in rnd.sample(ran80,1)) else
      "VIV_COL" if (3200<n<8200) and (True in rnd.sample(ran80,1)) else
      "VIV_COL" if (8200<=n>8200) and (True in rnd.sample(ran60,1)) else
      "OTROS"     
      for n in const_tot["superf_const"]]

## Aplicación al DF:
    
const_tot["tipolog"]=pd.Series(edif)

#Creamos una lista de colores según las tipologías de la columna "tipolog".

color= ['blue' if l == "OTROS" else 'green' if l== "VIV_AISL" else 'red' for l in const_tot["tipolog"]]

#Creamos una lista de tamaños de bolitas en relación a alguna variable:

size= [20 if n < 13 else 80 if n <50 else 180 if n <100 else 120
         for n in const_tot["duracion_sems"]]

#Ploteamos Scatter:
const_tot.plot.scatter(x='modul_precio', y='retrasos_sems', title="restrasos VS mod.Precio",
                       color=color, alpha=0.1,figsize=(13, 7),s=size)

const_tot.plot.scatter(x='superf_const', y='retrasos_sems', title="restrasos VS superf_const",
                       color=color, alpha=0.1,figsize=(13, 7),s=size)
const_tot.plot.scatter(x='duracion_sems', y='retrasos_sems', title="restrasos VS mod.Precio", alpha=0.1)
const_tot.plot.scatter(x='duracion_sems', y='modul_precio', title="restrasos VS mod.Precio", alpha=0.1)
```
