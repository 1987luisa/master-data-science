```python
linkdin_research = rcsv.drop(['TÉRMINOS DE BÚSQUEDA EN LINKEDIN', 'Column1'], axis=1)
linkdin_research = linkdin_research.rename(columns={"Column2": "BIM_Role",
                                                    "BÚSQUEDA EXACTA": "Exact_role",
                                                    'BÚSQUEDA NO EXACTA':'Unprecise_role'})
linkdin_research= linkdin_research.sort_values(['Exact_role'],ascending=1)
```

```python
def pairing(df,col1,col2):
    listofpairs=[]
    for i in df[col1]:
        listofpairs.append([i])
    for j,value in enumerate(df[col2],0):
        listofpairs[j].append(value)
    for i, value in enumerate(listofpairs):
        listofpairs[i]=tuple(value)
    tuplofpairs=tuple(listofpairs)
    return tuplofpairs

pairing(linkdin_research,'Exact_role','Unprecise_role')
        
```
